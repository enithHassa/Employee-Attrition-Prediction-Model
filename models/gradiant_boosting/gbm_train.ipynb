```python
import os
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
import joblib
import time
import json
```

```python
import os

# List files in the directory
directory_path = '/content/drive/My Drive/FDM Project Files/splits/'
try:
    files = os.listdir(directory_path)
    print(f"Files in '{directory_path}':")
    for file in files:
        print(file)
except FileNotFoundError:
    print(f"Error: The directory '{directory_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Paths
SPLITS_DIR = "/content/drive/My Drive/FDM Project Files/splits"
GBM_DIR = "/content/drive/My Drive/FDM Project Files/models/gradiant_boosting"
os.makedirs(GBM_DIR, exist_ok=True)

# Load train data
X_train = pd.read_csv(os.path.join(SPLITS_DIR, "X_train.csv"))
y_train = pd.read_csv(os.path.join(SPLITS_DIR, "y_train.csv")).squeeze()
```

```python
# Convert 'Overtime' column to numerical using one-hot encoding
X_train = pd.get_dummies(X_train, columns=['Overtime'], drop_first=True)
```

```python
# Save the column order so evaluation uses same columns
columns_path = os.path.join(GBM_DIR, "gbm_columns.json")
with open(columns_path, "w") as f:
    json.dump(list(X_train.columns), f)
```

```python
# Define model
gbm = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)
```

```python
# Train and time it
start_train = time.time()
gbm.fit(X_train, y_train)
end_train = time.time()
train_time = round(end_train - start_train, 3)

print(f" GBM trained successfully in {train_time} seconds")
```

```python
# Save model
model_path = os.path.join(GBM_DIR, "gbm_model.pkl")
joblib.dump(gbm, model_path)
print(" Model saved at:", model_path)

# Save training metadata
meta = {"Train_Time_sec": train_time}
with open(os.path.join(GBM_DIR, "gbm_meta.json"), "w") as f:
    json.dump(meta, f)
print(" Training metadata saved")
```
